```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
class image(object):
    def __init__(self, path, show_size=(3, 3)):
        self.img = cv2.imread(path)
        self.show_size = show_size
        
    def show(self, image_new=""):
        if len(image_new):
            image = image_new
        else:
            image = self.img
        
#         plt.figure(figsize=self.show_size)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # 转换为灰度
        plt.show()
    
    def median_blur(self, ksize=5):
        img = self.img
        img_new = cv2.medianBlur(img, ksize)
        self.show(image_new=img_new)
    
    def median_single(self, data, high, wide, ksize):   #image为传入灰度图像，ksize为滤波窗口大小
#         image = cv2.cvtColor(self.img, cv2.COLOR_RGB2GRAY)
#         high, wide = image.shape[:2]
        img = data.copy()
        mid = (ksize-1) // 2
        
        med_arry = []
        for i in range(high-ksize):
            for j in range(wide-ksize):
                for m1 in range(ksize):
                    for m2 in range(ksize):
                        med_arry.append(int(data[i+m1,j+m2]))

                med_arry.sort()   				#对窗口像素点排序
                img[i+mid,j+mid] = med_arry[(len(med_arry)+1) // 2]        #将滤波窗口的中值赋给滤波窗口中间的像素点
                del med_arry[:]
                
        return img
        
    def median(self, ksize=5):
        high, wide = self.img.shape[:2]
        B, G, R = cv2.split(self.img)
        b = self.median_single(B, high, wide, ksize)
        g = self.median_single(G, high, wide, ksize)
        r = self.median_single(R, high, wide, ksize)
        
        img = cv2.merge([b, g, r])  # 合并通道
        self.show(image_new=img)
        
```

```python
path = 'scarlett_johansson.jpg'
img = image(path=path)
img.show()
img.median_blur()
print("自定义 中值滤波def执行开始")
img.median()
print("自定义 中值滤波def执行结束")
```
